```python
#  nhận diện khuôn mặt với mtcnn trên một bức ảnh
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
 
# vẽ một hình ảnh với các đối tượng được phát hiện
def draw_image_with_boxes(filename, result_list):
    # tải ảnh lên
    data = pyplot.imread(filename)
    # vẽ hình ảnh
    pyplot.imshow(data)
    # Lấy bối cảnh cho các hộp vẽ
    ax = pyplot.gca()
    # vẽ mỗi hộp
    for result in result_list:
        # lấy tọa độ
        x, y, width, height = result['box']
        # tạo hình dạng
        rect = Rectangle((x, y), width, height, fill=False, color='red')
        # vẽ hộp
        ax.add_patch(rect)
        # vẽ các dấu chấm
        for key, value in result['keypoints'].items():
            # tạo và vẽ dấu chấm
            dot = Circle(value, radius=2, color='red')
            ax.add_patch(dot)
    # hiển thị ảnh
    pyplot.show()
 
# filename = r"C:\Users\Admin\Project\FaceDetectionDeeplearning\test1.jpg"
# tải hình ảnh từ file
pixels = pyplot.imread(filename)
# tạo model dò, sử dụng weights mặc định
detector = MTCNN()
# phát hiện khuôn mặt trong hình ảnh
faces = detector.detect_faces(pixels)
# Hiển thị các khuôn mặt trên hình ảnh gốc
draw_image_with_boxes(filename, faces)
```

```python
# Trích xuất và vẽ biểu đồ từng khuôn mặt được phát hiện trong một bức ảnh
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN
 
# vẽ các khuôn mặt tách biệt
def draw_faces(filename, result_list):
    # tải hình ảnh
    data = pyplot.imread(filename)
    # Vẽ mỗi khuôn mặt dưới dạng một subplot
    for i in range(len(result_list)):
        # lấy tọa độ
        x1, y1, width, height = result_list[i]['box']
        x2, y2 = x1 + width, y1 + height
        # xác định subplot
        pyplot.subplot(1, len(result_list), i+1)
        pyplot.axis('off')
        # vẽ mặt
        pyplot.imshow(data[y1:y2, x1:x2])
    # hiển thị
    pyplot.show()
 
# filename = 'test2.jpg'
# tải hình ảnh từ file
pixels = pyplot.imread(filename)
# tạo model dò, sử dụng weights mặc định
detector = MTCNN()
# phát hiện khuôn mặt trong hình ảnh
faces = detector.detect_faces(pixels)
# Hiển thị các khuôn mặt trên hình ảnh gốc
draw_faces(filename, faces)
```
